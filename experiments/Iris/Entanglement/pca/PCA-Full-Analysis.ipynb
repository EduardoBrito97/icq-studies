# Plotting graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_graph_mult_lines(list_of_xs, list_of_ys, labelX, labelY, list_of_xs_legend, title, step=1, ax=None, plot=True, y_ticks=0.1):
    colors = ["red", "black", "blue", "yellow", "gray", "green"]

    for i in range(len(list_of_xs)):
        if ax == None:
            plt.plot(list_of_xs[i][0::step], list_of_ys[i][0::step], color=colors[i], marker="o",  linestyle="-")
        else:
            ax.plot(list_of_xs[i][0::step], list_of_ys[i][0::step], color=colors[i], marker="o",  linestyle="-")
    if ax == None:
        ax = plt.gca()
    ax.tick_params(axis='y', colors='red')
    ax.tick_params(axis='x', colors='red')

    ax.set_xlabel(labelX)
    ax.xaxis.label.set_color('red')

    ax.set_ylabel(labelY)
    ax.yaxis.label.set_color('red')

    ax.set_title(title)
    ax.set_xticks(np.arange(0, 0.5, 0.05))
    ax.set_yticks(np.arange(0, max(10 * y_ticks, 0.6), y_ticks))

    if list_of_xs_legend != None:
        ax.legend([list_of_xs_legend[i] for i in range(len(list_of_xs))], loc="lower right", bbox_to_anchor=(1, -0.3))
    if plot:
        plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_graph_mult_lines(all_class_0_proposed, all_scores_proposed, "Negativity", "Accuracy", None, title="Proposed method - Class 0", step=1, ax=axs[0], plot=False, y_ticks=0.05)
plot_graph_mult_lines(all_class_1_proposed, all_scores_proposed, "Negativity", "Accuracy", None, title="Proposed method - Class 1", step=1, ax=axs[1], plot=False, y_ticks=0.05)
plot_graph_mult_lines(all_class_2_proposed, all_scores_proposed, "Negativity", "Accuracy", definitions_order, title="Proposed method - Class 2", step=1, ax=axs[2], plot=False, y_ticks=0.05)
plt.show()
```
