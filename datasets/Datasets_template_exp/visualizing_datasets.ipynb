```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
def visualize_dataset(dataset_name, normalize_db = False):
    dataset = np.load(dataset_name + ".npy")

    X,y = dataset[:,:-1],dataset[:,-1]

    if normalize_db:
        X = normalize(X, axis=0)

    X0 = X[y == 0]
    X1 = X[y == 1]

    title = dataset_name
    if normalize_db:
        title = "normalized " + title
        
    plt.plot(X0[:,0],X0[:,1],'o')
    plt.plot(X1[:,0],X1[:,1],'o')
    plt.title(title)
    plt.show()
```

```python
datasets = ["blobs", "circles", "ellipses", "linear_sep", "moons", "stripes", "xor"]
for dataset in datasets:
    visualize_dataset(dataset)
```

```python
datasets = ["blobs", "circles", "ellipses", "linear_sep", "moons", "stripes", "xor"]
for dataset in datasets:
    visualize_dataset(dataset, True)
```
